```python
import numpy as np
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq
import os
from pathlib import Path
import shutil
import cv2 
```

```python
train_images_0 = pq.read_table('berrybengali_preprocessed_0.parquet').to_pandas()
train_images_1 = pq.read_table('berrybengali_preprocessed_1.parquet').to_pandas()
train_images_2 = pq.read_table('berrybengali_preprocessed_2.parquet').to_pandas()
train_images_3 = pq.read_table('berrybengali_preprocessed_3.parquet').to_pandas()
test_images = pq.read_table('berrybengali_preprocessed_test.parquet').to_pandas()
```

```python
image_0_id = []
image_1_id = []
image_2_id = []
image_3_id = []
image_test_id = []
```

```python
for i in range(len(train_images_0)):
    image_0_id.append("Train_" + str(i))
    image_1_id.append("Train_" + str(i + 50210))
    image_2_id.append("Train_" + str(i + 100420))
    image_3_id.append("Train_" + str(i + 150630))
    
for i in range(len(test_images)):
    image_test_id.append("Test_" + str(i))
```

```python
train_images_0["image_id"] = image_0_id
train_images_1["image_id"] = image_1_id
train_images_2["image_id"] = image_2_id
train_images_3["image_id"] = image_3_id
test_images["image_id"] = image_test_id
```

```python
train0_from_pandas = pa.Table.from_pandas(train_images_0)
train1_from_pandas = pa.Table.from_pandas(train_images_1)
train2_from_pandas = pa.Table.from_pandas(train_images_2)
train3_from_pandas = pa.Table.from_pandas(train_images_3)
test_from_pandas = pa.Table.from_pandas(test_images)
```

```python
pq.write_table(train0_from_pandas, 'berrybengali_preprocessed_wid_0.parquet')
pq.write_table(train1_from_pandas, 'berrybengali_preprocessed_wid_1.parquet')
pq.write_table(train2_from_pandas, 'berrybengali_preprocessed_wid_2.parquet')
pq.write_table(train3_from_pandas, 'berrybengali_preprocessed_wid_3.parquet')
pq.write_table(test_from_pandas, 'berrybengali_preprocessed_wid_test.parquet')
```
