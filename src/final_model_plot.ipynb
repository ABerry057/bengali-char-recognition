```python
import json
import pandas as pd
import numpy as np
import seaborn as sns
import os
from pathlib import Path
```

```python
parent_dir = str(Path(os.getcwd()).parent)
```

```python
def plot_accuracy(json_path, output_name):
    with open(json_path) as f:
        model_history = json.load(f)
    modelhist_df = pd.DataFrame.from_dict(model_history)
    
    sns.set_style("darkgrid", {"axes.facecolor": ".9"})
    sns.set(rc={'figure.figsize':(11.7,8.27)})
    sns.set_context("notebook", font_scale=2.0, rc={"lines.linewidth": 2.5})
    ax1 = sns.lineplot(style="event", data=modelhist_df[['val_dense_consonant_accuracy', 'val_dense_root_accuracy', 'val_dense_vowel_accuracy']])
    ax1.set(xlabel='Number of Epochs', ylabel='Accuracy')
    ax1.figure.savefig(parent_dir+ "/figures/tuning_plots/"+str(output_name) + "_accuracy_plot.png")
```

```python
def plot_loss(json_path, output_name):
    with open(json_path) as f:
        model_history = json.load(f)
    modelhist_df = pd.DataFrame.from_dict(model_history)
    
    sns.set_style("darkgrid", {"axes.facecolor": ".9"})
    sns.set(rc={'figure.figsize':(11.7,8.27)})
    sns.set_context("notebook", font_scale=2.0, rc={"lines.linewidth": 2.5})
    ax2 = sns.lineplot(data=modelhist_df[['val_dense_consonant_loss', 'val_dense_root_loss', 'val_dense_vowel_loss']])
    ax2.set(xlabel='Number of Epochs', ylabel='Loss')
    ax2.figure.savefig(parent_dir+ "/figures/tuning_plots/"+str(output_name) + "_loss_plot.png")
```

```python
# Final Model
plot_accuracy(parent_dir + "/models/final_model.json", "final_model")
```
